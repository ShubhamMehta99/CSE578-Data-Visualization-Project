```
import re
import json
import numpy as np
import pandas as pd

rc = []
name = []
id = []
lng = []
pc = []
st = []
addr = []
lat = []

i = 0

for line in open('tempe_restaurants.json', 'r') :
    s = line
    #ff = line.split("}")
    #print(ff)
    k = s[s.find("{")+1:s.rfind("}")]
    n = k.split(',')
    #print(n)
    for j in n:
      
      arr = j.split(':')
      #print(arr)
      if arr[0] == ' "business_id"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        ns2 = ns1.split('"')
        id.append(ns2[1])
      elif arr[0] == ' "name"':
        #print(arr[1])
        k1 = arr[1].split('"')
        result = re.match(r".*", k1[1])
        name.append(result.group(0))
        
      elif arr[0] == ' "latitude"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        ns2 = ns1.split('"')
        lat.append(float(ns1))
      elif arr[0] == ' "longitude"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        ns2 = ns1.split('"')
        lng.append(float(ns1))
      elif arr[0] == ' "postal_code"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        ns2 = ns1.split('"')
        pc.append(int(ns2[1]))
      elif arr[0] == ' "address"':
        k1 = arr[1].split('"')
        result = re.match(r".*", k1[1])
        addr.append(result.group(0))  
      elif arr[0] == ' "stars"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        #print(ns)
        st.append(float(ns1))
      elif arr[0] == ' "review_count"':
        ns = arr[1].split(" ")
        ns1 = ns[1]
        ns2 = ns1.split('"')
        #print(ns2)


        rc.append(int(ns1))
      else:
        continue
    
    if i == 100:
      break


netList = []

for idd,na,ad,zi,la,lo,sta,rcc in zip(id,name,addr,pc,lat,lng,st,rc):
  netList.append([idd,na,ad,zi,la,lo,sta,rcc])


df = pd.DataFrame(netList,columns=['id','name','address','postal_code','latitude','longitude','star','review_count'])
print (df)

df.to_csv('Restaurants.csv', index=False) 


```
